## 1. Time Series Forecasting

```python
# Fetch data
fetcher = DataFetcher()
stock_data = fetcher.get_stock_data('AAPL', period='2y')
prices = stock_data['Close']

# Initialize forecaster
forecaster = TimeSeriesForecaster(model_type='random_forest')

# Fit model
forecaster.fit(prices, n_lags=20)

# Forecast
forecast = forecaster.predict(prices, n_periods=30)

print(f"Forecasted {len(forecast)} periods ahead")
print(forecast.head())
```
